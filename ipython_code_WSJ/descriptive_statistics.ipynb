```python
import pandas as pd
import numpy as np
from sklearn import cross_validation
from sklearn.metrics import label_ranking_average_precision_score
from sklearn.metrics import make_scorer
from VerbSelect import VerbSelectBetaMRREstimator, VerbSelectKdeMRREstimator, VerbSelectBaselineMRREstimator
from scipy import stats


# select and filter data
triples = pd.read_csv('triples.csv',names=['sub','verb','per'])
rising_verbs = triples[triples['verb'].isin(['rise','increase','grow',
'climb','jump','surge','gain','soar','raise','advance','boost'])]
falling_verbs = triples[triples['verb'].isin(['fall','decline','drop',
'slip','plunge','slide','lose','tumble','plummet','ease','decrease','reduce',
'dip','shrink'])]

rising_verbs.per = rising_verbs['per'].str.replace(' %,','')
rising_verbs.per = rising_verbs['per'].str.replace(' percent,','').astype('float64')
falling_verbs.per = falling_verbs['per'].str.replace(' %,','')
falling_verbs.per = falling_verbs['per'].str.replace(' percent,','').astype('float64')
rising_verbs = rising_verbs[rising_verbs.per < 100]
rising_verbs = rising_verbs[rising_verbs.per > 0]
falling_verbs = falling_verbs[falling_verbs.per < 100]
falling_verbs = falling_verbs[falling_verbs.per > 0]
```

```python
rising_verb_list = ['rise','increase','grow','climb','jump','surge','gain','soar','raise','advance','boost']
falling_verb_list = ['fall','decline','drop','slip','plunge','slide','lose','tumble','plummet','ease','decrease','reduce','dip','shrink']
```

```python
 for verb in rising_verb_list:
    print verb
    verb_list = rising_verbs[rising_verbs.verb == verb].per
    print "mean: "
    print verb_list.mean()
    print "std:  "
    print verb_list.std()
    print "skewness: "
    print stats.skew(verb_list)
    print "kurtosis: "
    print stats.kurtosis(verb_list)
    print "median: "
    print verb_list.median()
    print "IQR: "
    print verb_list.quantile(0.25)
    print verb_list.quantile(0.75)
    print "---------------------"
    
```
