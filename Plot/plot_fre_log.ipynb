```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set_style("white")
```

```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
falling_verbs = pd.read_csv('falling_verbs', header=None, names=['freq'], index_col=0)
df = falling_verbs
g = df[:10].plot.bar(ylim=[0,4500], yticks=[1000,2000,3000,4000])
plt.savefig('fig/prior_freq_fall.pdf')
```

```python
rising_verbs = pd.read_csv('rising_verbs',  header=None, names=['freq'], index_col=0)
df = rising_verbs
df[:10].plot.bar(ylim=[0,9000], yticks=[2000,4000,6000,8000])
plt.savefig('fig/prior_freq_rise.pdf')
```

```python
n = len(falling_verbs)
x = [np.log(i+1) for i in range(n)]
y = [np.log(falling_verbs.freq[i]) for i in range(n)]

df = pd.DataFrame({'log(rank)':x, 'log(freq)':y})
g = sns.regplot(x='log(rank)', y='log(freq)', data=df)
g.set(xlim=[-0.1,2.75], xticks=[0,0.5,1.0,1.5,2.0,2.5])
g.set(ylim=[3.75,9.5], yticks=[4,5,6,7,8,9])
plt.savefig('fig/prior_zipf_fall.pdf')
```

```python
n = len(rising_verbs)
x = [np.log(i+1) for i in range(n)]
y = [np.log(rising_verbs.freq[i]) for i in range(n)]

df = pd.DataFrame({'log(rank)':x, 'log(freq)':y})
g = sns.regplot(x='log(rank)', y='log(freq)', data=df)
g.set(xlim=[-0.1,2.75], xticks=[0,0.5,1.0,1.5,2.0,2.5]);
g.set(ylim=[3.75,9.5], yticks=[4,5,6,7,8,9]);
plt.savefig('fig/prior_zipf_rise.pdf')
```

```python
stats.powerlaw.fit(falling_verbs.freq), stats.powerlaw.fit(rising_verbs.freq)
```
